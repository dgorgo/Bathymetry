### Reshape the dataset

```python
lon_shape = data['lon'].shape
lat_shape = data['lat'].shape

# Subsample longitude to match the latitude shape
subsampled_lon = data['lon'][:lat_shape[0]]

# Create a new dataset with modified longitude
new_data = xr.Dataset({
    'lon': (['lat', 'lon'], data['lon'].values),
    'lat': (['lat'], data['lat'].values),
    'lon': (['lon'], subsampled_lon.values)
})

# Now, both 'lon' and 'lat' have the same shape in the new dataset
print("Longitude Shape:", new_data['lon'].shape)
print("Latitude Shape:", new_data['lat'].shape)

# Close the datasets when done
data.close()
new_data.close()
```

```python
mp = Basemap(projection = 'merc',
            llcrnrlon = 94.6472,
            llcrnrlat = -60.7164,
            urcrnrlon = -25.9216,
            urcrnrlat = 2.2931,
            resolution = 'i')
```

```python
subset_lon = Longitude[::100]  
subset_lat = Latitude[::100]  

lon, lat = np.meshgrid(subset_lon, subset_lat)
x, y = mp(lon, lat)

subset_depth = Depth[::100, ::100]  
c_scheme = mp.pcolor(x, y, np.squeeze(subset_depth), cmap='jet')

mp.drawcoastlines()
mp.drawcountries()
cbar = mp.colorbar(c_scheme, location='right', pad='10%')

plt.title('Sub-Ice Bathy')
plt.show()
```
